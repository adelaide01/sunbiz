```python
def get_next_characters(f, num):
    c = f.read(num)
    yield c.decode('utf-8')
```

```python
with open("data/florida_corporations_sample.txt", "rb") as f:
    for k,w in widths:
        for c in get_next_characters(f, w):
            print(k,w,":", c)
```
